# Statistics Concepts

## Population vs. Sample
1. Population is the entire group, sample is a subset of that group

## Normal Distribution
1. Bell shaped curve. Uses standard deviation. I.E,there is a 68% chance it will fall within 1 std, -1,1
2. Area is the same as the probability I.E, .34 between mean and -1 std

1. Use Integral A to B of e^-(x-u)^2/2(STD)^2/STD*SQRT(2PI) to find area or percentage between. 

## Standard Normal Distribution
1. When x is less than the mean, z scores are negative
2. Z score represeent std, z1=std1
3. Z = (x-u)/STD
4. X = u + z(STD)

## Measures of Central Tendency
1. Mean
2. Median
3. Mode

## Variance and Standard Deviation
1. Range max - min
2. Variance: Larger = more disperse, gives you arbitrary number. 
3. Std: SQRT(Variance)

## Correlation and Covariance

1. Covariance + is positively correleted, x inc, y inc
2. covariance - is negatively correlated, x dec, y dec
3. Correlation is used to make the units the same and get a range -1<p<1 
4. If 1, very strongly positive correlation, 
5. If -1, very strongly negative correlation

## Central Limit Theroem 
1. If we collect samples of size n from a population, calculate the mean of those samples, and plot on a histogram, the histogram will approiximate a normal distribution
2. Very useful because it says we don't care what the distribution is, the sample means will be normally distributed. 
### Law of Large Numbers
1. The more people you sample, the closer it will be to the mean.

## Hypothese Testing and The Null Hypothesis  
1. Have preliminary experiment, create a hypothesis, test the hypothesis.
2. If the other experiments aren't enogh evidence to support the hypothesis, we fail to reject the hypothesis
3. There hypothesis that there is no difference between things is the null hypothesis
4. Need a preliminary test without the null hypothesis
5. Assume there is no difference

1. Are distinguished between difference and correlation hypotheses as well as directed and undirected hypothese
2. Is there a difference between groups? Is there a correlation between variables?
3. Is there a difference or a correlation? Is there a difference or correlation in a certain direction?
4. Difference Hypothesis: The "group" of men earns more than the "group" of women. Smokers have a higher risk of heart attack than non-smokers.
5. Correlation Hypotheses: The bigger a person is, the heavier they are. The more horsepower a car has, the higher the consumption. Usually "the more" or "higher/lower"
6. Undirected Hypotheses (two-tailed): check if  there is a correlation or a difference, no matter in wich direction the correlation or the difference goes.
7. Ex: There is a difference between the salary of men and women. There is a relationship between height and weight of people.
8. Directed Hypotheses (one-tailed): Men got a higher salary than women. 

9. Hypothesis tests can only determine with a probability of error wheter a hypothesis is accepted or rejected. 

10. Types of error: 
11. Type 1: If the alternative hypothesis is accepted although the null hypothesis holds.
12. Type 2: When the null hypothesis is retained even though the alternative hypothesis holds

## P-Value testing
1. between values 0 and 1, quantify how confident we should be that Drug A is different from Drug B
2. The closer a p-value is to 0, the more confidence we have that they are different.
3. 0.05 is a common threshhold for no difference. 
4. This is hypothesis testing, the idea of trying to determine if these drugs are the same or not.
5. P-Value helps decide if we should reject the Null Hypothesis 
6. A small p-value helps us decide if Drug A is different from Drug B, it does not tell us how different they are. 

7. How likely it is to draw a sample in which the salary of men and women differs by more than 250 euros. 
8. If p-value is greater than signifigance level, we keep the null hypothesis, if it is less than or equal to the significance level, we reject the null hypothesis. 

## T Test
1. Asking a question about difference of means.
2. Two tailed t test to find a difference in any direction
3. One tailed t test to find a difference on one direction
4. Paired t test when there are pairs like countries from different surveys from different years. Can be one sided, two sided

## ANOVA
1. Difference of means but with 3 or more populations
2. Use tukey to see difference and p-value between

## Chi Squared
1. Categorial variables and proportions
2. Testing if there is a difference between the proportions between different categories 
3. Do we expect small, medium, and large amount to be equal? use goodness of fit test.
4. Does the size of the flower tell us anything about the species ? test of independence

```python
df = pd.read_csv("data/insurance.csv")
feature = "smoker"
label = "charges"
def anova(feature, label):
    groups = df[feature].unique()
    grouped_values = []
    for group in groups:
        grouped_values.append(df[df[feature] == group][label])
    return grouped_values

grouped_values = anova(feature, label)
stats.f_oneway(*grouped_values)
    
```

```python
df = pd.read_csv("http://www.ishelp.info/data/bikebuyers.csv")
df.head()
```

```python
df["Education_rank"] = df["Education"]
df["Education_rank"].replace(["Partial High School", "High School", "Partial College", "Bachelors", "Graduate Degree"],
                            [1,2,3,4,5], inplace=True)
```

```python
df["Commute_rank"] = df["Commute Distance"] 
df["Commute_rank"].replace(["0-1 Miles", "1-2 Miles", "2-5 Miles", "5-10 Miles", "10+ Miles"],
                         [0,1,2,5,10], inplace=True)
df["Commute_rank"]
```

```python
def clean_bike_buyers(df):
    df["Education_rank"] = df["Education"]
    df["Education_rank"].replace(["Partial High School", "High School", "Partial College", "Bachelors", "Graduate Degree"],
                                [1,2,3,4,5], inplace=True)
    df["Commute_rank"] = df["Commute Distance"] 
    df["Commute_rank"].replace(["0-1 Miles", "1-2 Miles", "2-5 Miles", "5-10 Miles", "10+ Miles"],
                             [0,1,2,5,10], inplace=True)
    df["Purchased Bike"].replace(["Yes", "No"], [0,1], inplace=True)
    
    return df
df = clean_bike_buyers(df)
```

```python
#Use tukey to print out a comparison between two groups

from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(df["Purchased Bike"], df["Education"])

print(mc.tukeyhsd())
```

```python
#Bonferoni 

e_types = df["Education"].unique()
ttests = []

for i,e in enumerate(e_types):
    for i2, e2 in enumerate(e_types[i+1:]):
        g1 = df[df["Education"] == e]["Purchased Bike"]
        g2 = df[df["Education"] == e2]["Purchased Bike"]
        t,p = stats.ttest_ind(g1,g2)
        ttests.append([f"{e} - {e2}", {t}, {p}])

threshold = 0.05/len(ttests)
ttests = [test for test in ttests if test[2] < threshold]
ttests
```
